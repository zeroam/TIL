```python
import re
import pandas as pd
```

```python
# 로그의 각 행에 매치하는 정규 표현
pattern = re.compile('^\S+ \S+ \S+ \[(.*)\] "(.*)" (\S+) (\S+)$')

# 정규 표현으로 파싱하는 함수(일치하지 않는 행은 그냥 버린다)
def parse_access_log(path):
    for line in open(path, 'rb'):
        line = line.decode(encoding='ascii', errors='ignore')
        for m in pattern.finditer(line):
            yield m.groups()
            
# 로그 파일을 읽어서 데이터 프레임으로 변환
columns = ['time', 'request', 'status', 'bytes']

pd.DataFrame(parse_access_log('access_log_Jul95'), columns=columns)
```

```python
# 데이터 프레임을 변수에 보관하기
df = pd.DataFrame(parse_access_log('access_log_Jul95'), columns=columns)
```

```python
# 'time' 칼럼을 덮어쓴다(타임 존 버리기)
df.time = pd.to_datetime(df.time, format='%d/%b/%Y:%X', exact=False)
```
